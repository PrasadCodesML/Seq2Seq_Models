### Sequence to Sequence Models

RNNs are used for data where sequence is followed or time series based data

Types of RNNs
1. Many to One
2. One to Many
3. Many to Many
  - Synchornous -> Input and output length is equal
  - Asynchornous -> Ouput and input length is not same -> This is the reason sequence to sequence models were created
  eg . Text summerization, Question answering, chatbot, speech to text, Language Translation

### 1. Encoder - Decoder Architecture

`Input -> Encoder -> State -> Decoder -> Output`

This Encoder Decoder architecture is useful for short term context but this is not useful for long inputs to solve this Attention Mechanism came into picture

### 2. Attention Mechanism

Youshua Bengio

This is also an Encoder Decoder architecture in previous method there was only one context vector but in Encoder Decoder architecture with attention there are multiple context vectors

Now since we have multiple hidden states for prediction of one word, here comes attention and tells which context or which hidden state is important or the best possible hidden state for the given input

But the problem was the computational complexity and training time was increased because for each word you have to calculate similarity with all the other words and the computational graphs was quadratic but researchers realized that the problem was not in attention but the problem was we were using LSTMs in encoder which takes input sequentially

### 3. Transformers

In this architecture Attention is all you need

Here there are encoder decoder blocks but LSTMs are not used and here you can input all the words in a sentence at once there is parallel processing due to which this became popular and training became fast

The only problem is you require GPUs but also it requires time but most importantly it requrires lots and lots of data to train transformers from scratch

### 4. Transfer Learning

Transfer Learning can also be used in NLP's domain earlier transfer learning was only used for CNN models

Why was it believed transfer learning cannot be used in domain of NLP
1. Task specificity : It was not possible to train a model who can do general tasks, initially they tried to do using Machine Translation but not good result
2. Lots of data

But in ULMFit -> They did language modelling (Next word prediction)

Why Language Modelling pretraining worked
1. Rich feature leanring -> It learns the grammer and basics of language which can be used in various tasks
2. Huge availabe of data -> Earlier for machine translation you need labelled data but in language modelling you don't need labelled data

ULMFit was trained on wikipedia data and the training was unspervised pretraining language modelling but in this transformers was not used

### LLMs

Google's BERT and OpenAI's GPT were trained on huge dataset and was very good at doing finetuning

BERT was Encoder only model and GPT was Decoder only model

But the parameters in these models were huge that's the reason they were called as Large Language Models

Qualities of LLMs
1. Training was done on very very large data and diverse data -> 45 Terabytes
2. Clusters of GPUs -> GPT3 was trained on super computer with 1000s of GPUs
3. Training Time -> Days to Weeks to train GPT3
4. Cost was very high
5. Energy
