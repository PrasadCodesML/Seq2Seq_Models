### Seq2Seq Data

1. Tabular Data -> ANN
2. Image Data -> CNN
3. Sequential Data -> RNNs {LSTMs, GRUs)
4. Sequence 2 Sequence data -> We have to solve this

### Encoder Decoder Architecture

Input is given to the Encoder and Encoder will try to capture the essence of the sentence and create a context vector, this context vector contains the summary and this context vector is inputted to Decoder the decoder then decodes the context vector and give the output

image.png

### What is inside Encoder

- Encoder consists of LSTMs or GRUs
image.png



- Decoder also consits of LSTMs or GRUs
The context vector of Encoder are given input to the Decoder along with a special symbol `<START>` and then the Decoder will start generating answer and this will happen until the decoder generates `<END>`

image.png

### Training of the architecture

Dataset -> Input (english) output (hindi)

image.png

1. Tokenizing the data
  - While encoding hindi you also have to encode `<START>` and `<END>`

2. Training
  - Forward Propagation
    - Initially the weights in LSTMs in Encoder and Decoder are randomly initialized
    - Before the output there is a softmax layer inside Decoder which generates probability of the words
    - While training decoder may output a different word at a timestamp and this word if sent further might generate wrong responses so even it a t=1 a different word was generated than expected at t=2 the correct words is sent this helps for faster conversion and this method is called as Teacher Forcing
  - Backward Propagation
    - We will calculate the loss (Categorical crossentropy)
    - Calculate Gradients
    - Update the weights based on the learning rate

3. Context Vector consists of h(t) and c(t)
