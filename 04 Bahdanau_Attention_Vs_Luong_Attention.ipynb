### Bahdanau Attention Vs Luong Attention

Bahdanau Attention

These two mechanism help us to find the weights for doing the weighted sum to calculate the Attention input also knows as alignment score

We need a function where we can input the s(i-1) (Previous state of decoder) and h(j) (hidden state of encoder) instead of creating a new function we will use ANN

This ANN is part of the Decoder

This ANN is known as alignment model and this Attention Mechanism is also known as additive attention

Luong Attention

Point of difference between Bahdanau Attention
1. Here to calculate the weights we use s(i) and h(j) [here we depend on current hidden state of the decoder]
2. Instead of a ANN we use dot project of s(i) Transpose and h(j)
3. Here or context vector is getting concatenated after decoder output


Here we do dot product of s(i) with each hidden state of encoder and then send it to softmax
