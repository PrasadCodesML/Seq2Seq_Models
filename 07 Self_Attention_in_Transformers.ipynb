### Points to consider

WhatsApp Image 2025-03-11 at 17.08.36_04fbbfe2.jpg


1. This operation of converting word embeddings into contextual embeddings is parallel process
2. There are no learning parameters involved

WhatsApp Image 2025-03-11 at 17.08.36_7a42b1e6.jpg

But this method by which we generated contextual embedding these are general contextual embeddings and not task specific contextual embeddings...

Why does it matter -> Because if let's say we are saying 'Piece of Cake' -> in hindi it has 2 meaning, but general contextual embeddings will give -> Cake ka tukda

To tackle this we have to introduce weights and biases in this process so that we can create task specific contextual embeddings

- The Green Boxes -> Query
- The Orange Boxes -> Keys
- The Blue Boxes -> Values

Each word embedding are playing 3 roles
1. Query
2. Key
3. Value

and instead of the same vector playing 3 roles what if we seperate this vector into 3 parts which can do their job properly because of Seperation of Concerns

So we will take our embedding vector and convert it into 3 vectors query, key and value and this query, key and value vectors will be created on basis of data


WhatsApp Image 2025-03-11 at 17.08.36_5815aa59.jpg

But how to we acutally create these query, keys and values -> We have to do linear transformations
we will multipy this embedding with 3 matrix Wq, Qk and Qv -> Initally this will be randomized but this will improve with help of backpropagation

The important point is Wq, Wk, and Wv will be same for all the words
