### Why do we need to scale Self Attention ?

dk -> Is dimension of key vector  (the key vector of all the words will have same dimensions)

key vector, query vector and value vector may have different dimensions

Why to we need to scale or divide by 1/sqrt(dk) -> This is because nature of dot product

When you have low dimension vectors and we calculate dot product then the variance is also low and when we calculate dot product with high dimensional vectors then we get high variance

In soft max if there is high variance between numbers then it will give higher number more probability and lower number very low probability due to which gap will increase, and while back propagation the focus will shift towards correcting those numbers and other numbers face vanishing gradient problems

The problem is when there are high dimensional embeddings the variance will also increase which will incraese the gap after softmax
