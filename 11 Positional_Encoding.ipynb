### Why we need Positonal Encoding

Since in self attention we are passing all the values parallelly self attention wouldn't know the sequence of the words, for self attention Prasad eats pizza and Pizza eats Prasad are both same because we pass them parallely

This is the reason we need Positional Encoding which tells what is the sequence of the words

How will the positional embedding be created for a word : It will be created using sine and cosine curve -> we will take a sine curve and a cosine curve and then place them in first and second position and then we will take another sine and cosine curve and then add them at third and fourth place and this will go on... untill it matches the dimension of word embedding
