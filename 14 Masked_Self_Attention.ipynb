### Autoregressive Models

The transformer decoder is auto regressive at inference time and non-autoregressive at training time

But why the reason is Masked Self Attention

Inference : Prediction

Auto regressive : In context of deep learning, auto regressive models are a class of model that generate data points in a sequence by conditioning each new point on the previously generated points

The transformer decoder is non auto regressive at training time because we have a concept called teacher forcing, when there is output to be generated from first decoder block, it is not necessary to send the generated output to send as input to the next decoder block because we have the dataset we can pass the correct values and parallelize the generation of outputs of decoder

The transformer decoder at inference has to be auto regressive because there is no choice, we have to use the previous input since we don't have the dataset
