### What is Cross Attention

Cross-attention is a mechanism used in transformer architectures, particularly in tasks involving sequence-to-sequence data like translation or summarization. It allows a model to focus on different parts of an input sequence when generating an output sequence.

Self Attention is the way to find out similarity between words in same sentence but Cross Attention is the way to find out similarity between two different sentences

In the architecture you can see there are some inputs to the decoder which are coming from the encoder

image.png

We need cross attention because there are only two things on which the next word prediction will depend on
1. What is being predicted uptill now -> Self attention
2. What is in the input sentence -> Cross attention

Cross Attention is conceptually very similar to Self attention

## Self Attention Vs Cross Attention

1. The input

Self Attention : Embedding of a single sequence or contextual embedding of previous self attention layer

Cross Attention : Embedding of single sequence or contextual embedding of previous layer of self attention and contextual embedding of predicted values uptill now

image.png

2. The biggest difference between Cross attention and self attention lies in calculation of query, key and value vector

In this case the query vector is calculated using decoders predicted words and from the words from the input sequence we calculate the key and value vector and further work as all the same

3. The Output

The no of contextual embeddings are same as the no of words in the output sequence
