### Transformer Inference

1. Encoder -> The encoder of transformer behaves the same during inference as it does during training

Tokenization -> Positional Encoding -> (Multi head attention -> ADD and Normalization -> Feed Forward Neural Network -> ADD and Normalization) * 6

2. Decoder behave differently because we do teacher forcing during training so it acts as non auto regressive and behaves as auto regressive during inference

At the start of the Decoder you send a <sos> token (start of sentence) as the decoder knows about it it gets to know to start the process

Now the embedding layer will tokenize the <sos> token and then we will pass to the positional encoding and then this is sent to the decoder, and then it is sent to Masked Multi head attention block, which contain 3 matrix to get query, key and value vectors and then multiply query and key to get attention score and then dot product with value vector and get contextual embedding and then we do addition with residual connection and then send this to the normalization layer and then we send this to cross attention with outputs of the encoder which contains query matrix, value matrix, and key matrix we will multiply this with the query matrix we will get the query vector, and the key matrix and the value matrix will multiply and generate the query and key vectors for all the inputs given to the encoder and now we will multiply this query vector with all the key vectors and we will get the weights which we will multiply them with the value vector and concatenate and this will be output for the cross attention and then we will add it with z1Norm and then we will pass to layer normalization and then we will get zC1Norm which we will pass to the Feed Forward Neural Network and then again Add and Normalization layer and then this output will be sent to other 5 decoder blocks after which there is a linear layer with softmax with which we will get final output.
